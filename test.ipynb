```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 画像を読み込む
img = mpimg.imread('examples/data/example.png')

# ファイルを開いてデータを読み込む
with open("examples/data/example.txt", "r") as file:
    list_of_lists = [list(map(float, line.split())) for line in file]

# 描画の設定
fig, ax = plt.subplots()
ax.imshow(img)  # 画像をプロット

# ファイルから読み込んだデータを元にポリゴンを描画
for points in list_of_lists:
    label = points[0]  # ラベル（使用しないが読み込み）
    coords = points[1:]
    x_coords = [x * img.shape[1] for x in coords[0::2]]  # 正規化されたX座標をピクセル座標に変換
    y_coords = [y * img.shape[0] for y in coords[1::2]]  # 正規化されたY座標をピクセル座標に変換
    # ポリゴンをプロット（初点を終点に追加して閉じる）
    ax.plot(x_coords + [x_coords[0]], y_coords + [y_coords[0]], marker='o', linestyle='-', color='red')

# グラフの設定
ax.set_title("Polygon on Image")
ax.set_xlim([0, img.shape[1]])
ax.set_ylim([img.shape[0], 0])
ax.set_xlabel("X coordinate")
ax.set_ylabel("Y coordinate")
ax.set_aspect('equal')  # 画像のアスペクト比に合わせる

# グラフを表示
plt.show()
```

```python
def calculate_polygon_area(coords):
    """ポリゴンの面積を計算する関数。coordsは(x1, y1, x2, y2, ..., xn, yn)の形式のリスト。"""
    n = len(coords) // 2  # 座標点の数
    area = 0
    for i in range(n):
        x1, y1 = coords[2 * i], coords[2 * i + 1]
        if i == n - 1:  # 最後の点の次は最初の点
            x2, y2 = coords[0], coords[1]
        else:
            x2, y2 = coords[2 * (i + 1)], coords[2 * (i + 1) + 1]
        area += x1 * y2 - y1 * x2
    return abs(area) / 2

# 使用例
# ファイルからデータを読み込む
with open("examples/data/example.txt", "r") as file:
    polygons = [list(map(float, line.split())) for line in file]

# 各ポリゴンの面積を計算
for points in polygons:
    coords = points[1:]  # ラベルを除外
    x_coords = [x * img.shape[1] for x in coords[0::2]]  # 正規化解除
    y_coords = [y * img.shape[0] for y in coords[1::2]]
    flat_coords = [coord for pair in zip(x_coords, y_coords) for coord in pair]
    area = calculate_polygon_area(flat_coords)
    print(f"Polygon area: {area:.2f} square units")
```

```python
import matplotlib.image as mpimg

# 画像の読み込み
img = mpimg.imread('examples/data/example.png')
height, width = img.shape[0], img.shape[1]

# 画像の四隅の座標を設定
coords = [0, 0, width, 0, width, height, 0, height]

def calculate_polygon_area(coords):
    """シューレース公式を用いて多角形の面積を計算"""
    n = len(coords) // 2
    area = 0
    for i in range(n):
        x1, y1 = coords[2 * i], coords[2 * i + 1]
        # 修正: 条件式の結果をタプルで囲む
        x2, y2 = (coords[0], coords[1]) if i == n - 1 else (coords[2 * (i + 1)], coords[2 * (i + 1) + 1])
        area += (x1 * y2 - y1 * x2)
    return abs(area) / 2

# 四隅の座標に基づいて面積を計算
calculated_area = calculate_polygon_area(coords)
actual_area = width * height

print(f"Calculated area using Shoelace formula: {calculated_area}")
print(f"Actual area based on image dimensions: {actual_area}")
```
